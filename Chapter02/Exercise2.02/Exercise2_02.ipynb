```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('circles.csv')
plt.scatter(data.iloc[:,0], data.iloc[:,1])
plt.show()
```

```python
plt.hist(data.iloc[:,0])
plt.show()
```

## Implementing k-Means

```python
from sklearn.cluster import KMeans
```

```python
ideal_k = []

for i in range(1, 21):
    est_kmeans = KMeans(n_clusters=i, random_state=0)
    est_kmeans.fit(data)
    
    ideal_k.append([i, est_kmeans.inertia_])
```

```python
ideal_k = np.array(ideal_k)
```

```python
plt.plot(ideal_k[:,0], ideal_k[:,1])
plt.show()
```

Based on the graph we will select 5 clusters

```python
eat_kmeans = KMeans(n_clusters=5, random_state=0)
est_kmeans.fit(data)
pred_kmeans = est_kmeans.predict(data)
```

```python
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=pred_kmeans)
plt.show()
```
